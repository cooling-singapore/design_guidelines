```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import matplotlib
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

from matplotlib.backends.backend_pdf import PdfPages
%matplotlib inline
```

```python
# LOCAL VARIABLES

# POINTERS
DATA_FOLDER_PATH = os.path.join(os.getcwd(),'input')
OUTPUTDATA_FOLDER_PATH = os.path.join(os.getcwd(),'output', 'scatterplot')

# SCENARIOS
SCENARIOS_LIST = list(range(57)) + [61] + list(range(63, 67)) #67
VARIABLES_LIST = ["pet"]
WORKSHEET_CASE1 = ['Sheet1', 'Sheet2']
WORKSHEET_CASE1_WEIGHTS = [0.0, 1.0]
WORKSHEET_CASE2 = ['Sheet1', 'Sheet2', 'Sheet3']
WORKSHEET_CASE2_WEIGHTS = [0.0, 0.5, 0.5]

# COLOR MAPS
BLUE = (0.164705882, 0.674509804, 0.88627451, 1) # 42, 172, 226
BLACK = (0,0,0, 1) # 0,0,0
BLUE_LIGHT = (0.564705882, 0.847058824, 0.941176471, 1) #144, 216, 240
BLUE_FILL = (0.164705882, 0.674509804, 0.88627451, 0.2) #42, 172, 226, 20%
GREY = (0.850980392, 0.850980392, 0.850980392, 1) #217,217 ,217

MIN_VALUE_PET = 25
MAX_VALUE_PET = 45

#SIZE FIGURE
SIZE_X = 11.4
SIZE_Y = 3.515
```

```python
# DO SOME PROCESSING HERE
MIN_DICT = {
            "pet":MIN_VALUE_PET
            }

MAX_DICT = {
            "pet":MAX_VALUE_PET,
            }
```

```python
for scenario in SCENARIOS_LIST:
    for variable_name in VARIABLES_LIST:
        print("Evaluating scenario {} and variable {}".format(scenario, variable_name))
        input_file_path = os.path.join(DATA_FOLDER_PATH, variable_name, "Scenario_"+str(scenario)+".xls")
        
        # read what sheets are there
        xl = pd.ExcelFile(input_file_path)
        sheet_names = xl.sheet_names
        
        # Get weighted average 
        counter = 0
        if sheet_names == WORKSHEET_CASE1:
            # run first case of weighted averages
            for worksheet, weight in zip(WORKSHEET_CASE1, WORKSHEET_CASE1_WEIGHTS):
                if counter == 0:
                    weighted_df = pd.read_excel(input_file_path, sheet_name=worksheet, header=None) * weight
                else:
                    weighted_df += pd.read_excel(input_file_path, sheet_name=worksheet, header=None) * weight
                counter +=1
        
        elif sheet_names == WORKSHEET_CASE2:
            # run first case of weighted averages
            for worksheet, weight in zip(WORKSHEET_CASE2, WORKSHEET_CASE2_WEIGHTS):
                if counter == 0:
                    weighted_df = pd.read_excel(input_file_path, sheet_name=worksheet, header=None) * weight
                else:
                    weighted_df += pd.read_excel(input_file_path, sheet_name=worksheet, header=None) * weight
                counter +=1
                
        # calculate mean and variances
        mean = weighted_df.iloc[3][:24]
        positive_variance = mean + np.sqrt(weighted_df.iloc[11][:24])
        negative_variance = mean - np.sqrt(weighted_df.iloc[11][:24])
        
        # get figure title
        maximum = weighted_df.max().max()
        minimum = weighted_df.min().min()
        title = "The maximum was: {}, the minimum was: {}".format(round(maximum,2), round(minimum,2))

        # get figure size
        figsizex = SIZE_X + 2.5
        figsizey = SIZE_Y + 1.5 #this is addded extra
        
        #create pdf
        output_file_path = os.path.join(OUTPUTDATA_FOLDER_PATH, "S"+str(scenario)+"_scatterplot_"+variable_name+".pdf")
        pdf_pages = PdfPages(output_file_path)

        # fill in pdf with the figure
        fig, ax = plt.subplots(1, 1, figsize=(figsizex,figsizey))
        ax.plot(mean, color=BLACK)
        ax.plot(positive_variance, color=BLUE)
        ax.plot(negative_variance, color=BLUE_LIGHT)
        ax.set_xticks(np.arange(0, 24, 1))
        ax.set_yticks(np.arange(MIN_DICT[variable_name], MAX_DICT[variable_name], 1))
        ax.yaxis.grid(color=GREY, linestyle='-', linewidth=0.25)
        ax.xaxis.grid(color=GREY, linestyle='--', linewidth=0.25)
        ax.set_title(title)
        ax.fill_between(negative_variance.index, negative_variance, positive_variance, color=BLUE_FILL)
        plt.ylim(MIN_DICT[variable_name], MAX_DICT[variable_name]) 
        plt.xlim(0, 23) 
        ax.set_frame_on(False)
        #plt.show()

        #save figure, close figure and pdf (saves memory)
        pdf_pages.savefig(fig)
        plt.close()
        pdf_pages.close()
```
